```python
from sklearn import datasets, model_selection

import numpy as np
```

### Разовое разбиение данных на обучение и тест с помощью train_test_split

```python
iris = datasets.load_iris()
```

```python
train_data, test_data, train_labels, test_labels = model_selection.train_test_split(iris.data, iris.target, 
                                                                                     test_size = 0.3)
```

```python
#убедимся, что тестовая выборка действительно составляет 0.3 от всех данных

float(len(test_labels))/len(iris.data)
```

```python
print('Размер обучающей выборки: {} объектов \nРазмер тестовой выборки: {} объектов'.format(len(train_data),
                                                                                            len(test_data)))
```

```python
print ('Обучающая выборка:\n', train_data[:5])
print ('\n')
print ('Тестовая выборка:\n', test_data[:5])
```

```python
print( 'Метки классов на обучающей выборке:\n', train_labels)
print( '\n')
print( 'Метки классов на тестовой выборке:\n', test_labels)
```

### Стратегии проведения кросс-валидации

#### KFold

```python
for train_indices, test_indices in model_selection.KFold(n_splits = 5).split(iris['data']):
    print( train_indices, test_indices)
```

```python
for train_indices, test_indices in  model_selection.KFold(n_splits = 2, shuffle = True).split(iris['data']):
    print( train_indices, test_indices)
```

```python
for train_indices, test_indices in  model_selection.KFold(n_splits=2, shuffle = True, random_state = 1).split(iris['data']):
    print( train_indices, test_indices)
```

#### StratifiedKFold

```python
for train_indices, test_indices in model_selection.StratifiedKFold(n_splits = 2, shuffle = True, random_state = 0).split(X=iris['data'],y=iris['target']):
    print( train_indices, test_indices)
```

#### ShuffleSplit

```python
for train_indices, test_indices in model_selection.ShuffleSplit(n_splits = 10, test_size = 0.2).split(iris['data']):
    print( train_indices, test_indices)
```

```python
for train_indices, test_indices in model_selection.StratifiedShuffleSplit(n_splits=4, test_size = 0.2).split(iris['data'], iris['target']):
    print (train_indices, test_indices)
```

#### Leave-One-Out

```python
for train_indices, test_index in model_selection.LeaveOneOut().split(iris['data']):
    print( train_indices, test_index)
```
