```python
import pandas as pd
import csv

with open('checkins.dat') as input_file:        
    newLines = []
    for line in input_file:
        newLine = [x.strip() for x in line.split('|')]
        if len(newLine) == 6 and newLine[3] and newLine[4]:
            newLines.append(newLine)
```

```python
df = pd.DataFrame(newLines[1:], columns=newLines[0])
df.to_csv('file.csv', sep = '\t')
df_shorten = pd.DataFrame(newLines[1:100001], columns=newLines[0])
df_c = df_shorten.ix[:,['latitude', 'longitude']]
```

```python
df_c.latitude = df_c.latitude.apply(float)
df_c.longitude = df_c.longitude.apply(float)

```

```python
df_c[(df_c.latitude>-33.87)&(df_c.latitude<-33.86)&(df_c.longitude <151.3)&(df_c.longitude >151.1)]
```

```python
from sklearn import cluster
clst = cluster.MeanShift(bandwidth=0.1)
clst.fit(df_c)
```

```python
k = pd.DataFrame(clst.cluster_centers_)
#k = clst.cluster_centers_
```

```python
import numpy as np
d = clst.labels_
unique, counts = np.unique(d, return_counts=True)
list_ind = list()
for i in (np.asarray((unique, counts)).T):
    if i[1]>=15:
        list_ind.append(i[0])
```

```python
more_15 = k[list_ind]#624 indecis
more_15_centers = pd.DataFrame(more_15, columns = ['latitude', 'longitude'])
```
