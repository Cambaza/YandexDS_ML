```python
df['nevermarr'] = df['agefm'].isnull().astype('int')
```

```python
cols = ['ceb', 'age', 'educ', 'religion', 'idlnchld', 'knowmeth', 'usemeth',
       'evermarr', 'agefm', 'heduc', 'urban', 'electric', 'radio', 'tv',
       'bicycle', 'nevermarr']
```

```python
col1 = ['ceb', 'age', 'educ', 'religion', 'idlnchld', 'knowmeth', 'usemeth', 'agefm', 'heduc', 'urban', 'electric', 'radio', 'tv',
       'bicycle', 'nevermarr']
```

```python
df1 = df[col1]
```

```python
df1['agefm'].fillna(0, inplace=True)
```

```python
df1['heduc'] = df1.apply(lambda x: -1 if (pd.isnull(x['heduc']) and x['nevermarr']==1) else x['heduc'], axis=1)
```

```python
sum(df1['heduc'].isnull())
```

```python
df1['idlnchld_noans'] = df['idlnchld'].isnull().astype('int')
df1['heduc_noans'] = df['heduc'].isnull().astype('int')
df1['usemeth_noans'] = df['usemeth'].isnull().astype('int')
```

```python
df1['idlnchld'].fillna(-1, inplace=True)
df1['heduc'].fillna(-2, inplace=True)
df1['usemeth'].fillna(-1, inplace=True)
```

```python
import statsmodels.formula.api as smf
```

```python
age + educ + religion + idlnchld + knowmeth + usemeth +'\
'agefm + heduc + urban + electric  + bicycle '\
'nevermarr + idlnchld_noans + heduc_noans + usemeth_noans'
```

```python
m1 = smf.ols('ceb ~ age + educ + religion + idlnchld + knowmeth + usemeth +'\
'agefm + heduc + urban + electric + radio + tv + bicycle +'\
'nevermarr + idlnchld_noans + heduc_noans + usemeth_noans', 
             data=df1)
fitted = m1.fit()
print(fitted.summary())
```

```python
import statsmodels.stats.api as sms


print('Breusch-Pagan test: p=%f' % sms.het_breuschpagan(fitted.resid, fitted.model.exog)[1])
```

```python
m2 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + usemeth +'\
'agefm + heduc + urban + electric + bicycle +'\
'nevermarr + idlnchld_noans + heduc_noans + usemeth_noans', 
             data=df1)
fitted = m2.fit(cov_type='HC1')
print(fitted.summary())
```

```python

print('Breusch-Pagan test: p=%f' % sms.het_breuschpagan(fitted.resid, fitted.model.exog)[1])
```

```python
print("F=%f, p=%f, k1=%f" % m1.fit().compare_f_test(m2.fit()))
```

```python
m3 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth +'\
'agefm + heduc + urban + electric + bicycle +'\
'nevermarr + idlnchld_noans + heduc_noans', 
             data=df1)
fitted = m3.fit(cov_type='HC1')
print(fitted.summary())
```
